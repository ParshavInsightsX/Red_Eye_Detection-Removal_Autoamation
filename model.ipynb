```python
import pandas as pd
import os
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from keras.preprocessing.image import ImageDataGenerator
```

```python
directory_path = '/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/eye_data'
directory_files = os.listdir(directory_path)
```

```python
directory_files
```

```python
x=[]
y=[]

for file_type in directory_files:
    img_folder_path=os.path.join(directory_path, file_type)
    img_name_list=os.listdir(img_folder_path)
    # img_path=os.path.join()
    for img_file in img_name_list:
      image_path=os.path.join(img_folder_path,img_file)
      img=cv2.imread(image_path)
      x.append(img)
      y.append(int(file_type))
```

```python
dim=(32,32)
dim
```

```python
X=[]
resized_list=[]
for image in x:
  resized=cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
  X.append(resized)
  resized_list.append(resized.shape)
```

```python
x_aug=[]
y_aug=[]


# Define the augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.08,
    height_shift_range=0.08,
    horizontal_flip=True,
    fill_mode='nearest')

index=0
for i in X:
  auggen_img_count=0
  label=y[index]
  index=index+1
  temp_img=i.reshape((1,)+i.shape)
  for batch in datagen.flow(temp_img, batch_size=1, save_to_dir='/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/aug_data', save_format='jpeg'):
    batch=np.squeeze(batch)
    x_aug.append(batch)
    y_aug.append(label)
    auggen_img_count +=1
    if auggen_img_count>10:
      break
```

```python
X=np.array(X)
y=np.array(y)

x_aug=np.array(x_aug)
y_aug=np.array(y_aug)
```

```python
X.shape,y.shape
```

```python
x_aug.shape,y_aug.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_aug,y_aug,test_size=0.2,random_state=104,shuffle=True)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.utils import plot_model
```

```python
img_shape=x_train[0].shape
img_shape
```

```python
model =Sequential()
model.add(Conv2D(32,(3,3), activation='relu', input_shape=img_shape))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(48,(3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(64,(3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(300, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
x=model.fit(x_train, y_train, epochs=20, batch_size=32, verbose=2, validation_split=0.1)
```

```python
save_model=False
path='/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/model.h5'
if save_model==True:
  model.save(path)
```

```python
loss, accuracy = model.evaluate(x_test,y_test, verbose=1)
print(f'Accuracy: {accuracy*100}')
```
