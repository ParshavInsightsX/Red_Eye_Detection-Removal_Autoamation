# Function eye_finder() to detect eyes and and crop that area of image


```python
def eye_finder(img):
  return_eyes=[]
  detector = dlib.get_frontal_face_detector()
  predictor = dlib.shape_predictor('/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/shape_predictor_68_face_landmarks.dat')
  print("original image dimension ",img.shape)
  if img.shape[0]>600 and img.shape[1]>600:  
    img=cv2.resize(img,None,fx=0.40,fy=0.40)
    print("new resize image dimension ",img.shape)
  real_img=img.copy()

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  rects = detector(gray, 1)
  shape_list=[]
  def shape_to_np(shape, dtype="int"):
      coords = np.zeros((68, 2), dtype=dtype)
      for i in range(0, 68):
          coords[i] = (shape.part(i).x, shape.part(i).y)
      return coords
  for (i, rect) in enumerate(rects):
      shape = predictor(gray, rect)
      shape = shape_to_np(shape)
      shape_list.append(shape)
  print("No of faces detected : ",len(shape_list),"\n\n")

  for shape in shape_list:
    ## Getting eyes points from predictor shape by model _______________________X_________________________________________________________

    ## left = [36, 37, 38, 39, 40, 41]
    ## right = [42, 43, 44, 45, 46, 47]

    left_eye=[]
    right_eye=[]

    left_eye=shape[36:42]
    right_eye=shape[42:48]

    ## Eye coordinate calculation in image  _________________________________________________X_____________________________________________

    left_x=1000
    left_y=1000
    left_w=-1
    left_h=-1

    right_x=1000
    right_y=1000
    right_w=-1
    right_h=-1

    for i in range(6):
      left_x=min(left_x,left_eye[i][0])
      left_y=min(left_y,left_eye[i][1])
      left_w=max(left_w,left_eye[i][0])
      left_h=max(left_h,left_eye[i][1])
      
      right_x=min(right_x,right_eye[i][0])
      right_y=min(right_y,right_eye[i][1])
      right_w=max(right_w,right_eye[i][0])
      right_h=max(right_h,right_eye[i][1])

    # print(left_x, left_y, left_w, left_h)
    # print(right_x, right_y, right_w, right_h)

    ## Eyes cropping from Real image  _________________________________________________X_____________________________________________
    extra=3

    left_start_point=(left_x-extra, left_y-extra)
    left_end_point=(left_w+extra, left_h+extra)

    right_start_point=(right_x-extra, right_y-extra)
    right_end_point=(right_w+extra, right_h+extra)

    left_eye_box = real_img[left_y-extra:left_h+extra , left_x-extra:left_w+extra]
    right_eye_box = real_img[right_y-extra:right_h+extra , right_x-extra:right_w+extra]

    ## Plotting of images of eyes and temporary full eyes box image _________________________________________________X_____________________________________________

    # color=(255, 0, 0)
    # thickness=1
    # demo_img = cv2.rectangle(img, left_start_point, left_end_point, color, thickness)
    # demo_img = cv2.rectangle(img, right_start_point, right_end_point, color, thickness)
    # cv2_imshow(demo_img)
    
    ## resizing the eyes image  to 32 X 32

    dim=(32,32)
    left_inlarge_img=cv2.resize(left_eye_box,dim,interpolation=cv2.INTER_AREA)
    right_inlarge_img=cv2.resize(right_eye_box,dim,interpolation=cv2.INTER_AREA)

    cv2_imshow(left_inlarge_img)
    print("\n")
    cv2_imshow(right_inlarge_img)

    return_eyes.append(left_inlarge_img)
    return_eyes.append(right_inlarge_img)

    print("\n One face process finished ________________________X______________________________________\n")
  return return_eyes
```

```python
from PIL import Image
import os
from pathlib import Path
import random


eyes_images=[]
eyes_label=[]
y_label_count=0

# eye_image_directory="/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/eye_data"
# os.chdir(eye_image_directory)


for image in x:
  cv2_imshow(image)
  print("\n y_label_count ",y[y_label_count],"\n")
  process_list = eye_finder(image)
  for i in process_list:
    eyes_images.append(i)
    eyes_label.append(y[y_label_count])
    if y[y_label_count]==0:
      eye_image_directory="/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/eye_data/0"
      os.chdir(eye_image_directory)
      filename=str(random.randint(0,999999))+".jpeg"
      cv2.imwrite(filename, i)
    else:
      eye_image_directory="/content/drive/MyDrive/Colab Notebooks/13th eye $ face detection/projectX/eye_data/1"
      os.chdir(eye_image_directory)
      filename=str(random.randint(0,999999))+".jpeg"
      cv2.imwrite(filename, i)
  y_label_count=y_label_count+1
```
